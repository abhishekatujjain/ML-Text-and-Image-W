# Image Data Preprocessing

```python
from skimage.io import imread, imshow
```

```python
peacock_img_color = imread("peacock.jpg", as_gray = False)
print("The shape of the colored image is ", peacock_img_color.shape)
print(peacock_img_color)
imshow(peacock_img_color)
```

```python
#reading image in gray color

peacock_img_gray = imread("peacock.jpg", as_gray= True)
print('shape of gray image', peacock_img_gray.shape)
imshow(peacock_img_gray)
```

```python
#showing image in different colors

import matplotlib.pyplot as plt
from skimage.color import rgb2gray, rgb2hsv
peacock = imread('peacock.jpg')
peacock_gray = rgb2gray(peacock)
peacock_hsv = rgb2hsv(peacock)

#displaying image
plt.figure(1, figsize = (30,10))
plt.subplot(131)
imshow(peacock)
plt.title('Original RGB Format')
plt.subplot(132)
imshow(peacock_gray)
plt.title('Gray colored')
plt.subplot(133)
imshow(peacock_hsv)
plt.title('HSV colored')
plt.show()
```

# image resizing and rescaling

```python
from skimage.transform import resize, rescale

print("Original Dimension", peacock.shape)

#resize 
pk_resize = resize(peacock, (120,120))
print('Resized image', pk_resize.shape)

#rescaling for gray image, rescaling image at 50%
pk_rescaled = rescale(peacock_gray, scale = (0.50))
print('Rescaled image', pk_rescaled.shape)

#New rescaling
pk_rescaled_new = rescale(peacock_gray, scale = (0.75))
print('Rescale dimension 75%', pk_rescaled_new.shape)

# Plot
plt.figure(1, figsize = (10,10))
plt.subplot(221)
imshow(peacock)
plt.title("Original One")

plt.subplot(222)
imshow(pk_resize)
plt.title("Resized")

plt.subplot(223)
imshow(pk_rescaled)
plt.title("Rescaled 50%")

plt.subplot(224)
imshow(pk_rescaled_new)
plt.title("Rescaled 75%")

plt.show()
```

# Image Rotation

```python
from skimage.transform import rotate

#rotating the image by 40 degress

pk_rotated = rotate(peacock, angle =40)
imshow(pk_rotated)
```

```python
from numpy import fliplr, flipud

#flipping image from left to right
pk_fliplr = fliplr(peacock)

#flipping image from up to down
pk_flipud = flipud(peacock)

#displaying image
plt.figure(1, figsize = (10,15))

plt.subplot(131)
imshow(peacock)
plt.title("Original")

plt.subplot(132)
imshow(pk_fliplr)
plt.title('Flipping left to right')

plt.subplot(133)
imshow(pk_flipud)
plt.title('Flipping up down')

plt.show()
```

# Image Intensity

```python
from skimage import exposure

#Adjusting brightness of image
pk_bright = exposure.adjust_gamma(peacock, gamma=0.5, gain =1)
pk_brightest = exposure.adjust_gamma(peacock, gamma = 0.25, gain =1)
pk_dark = exposure.adjust_gamma(peacock, gamma = 2.25, gain =1)
pk_darkest = exposure.adjust_gamma(peacock, gamma = 4, gain =1)

#displaying image
plt.figure(1, figsize = (20,15))

plt.subplot(131)
imshow(peacock)
plt.title("Original")

plt.subplot(132)
imshow(pk_bright)
plt.title('Brighter Image')

plt.subplot(133)
imshow(pk_brightest)
plt.title('Brightest Image')

plt.show()


#displaying image in darker tone
plt.figure(1, figsize = (20,20))

plt.subplot(131)
imshow(peacock)
plt.title("Original")

plt.subplot(132)
imshow(pk_dark)
plt.title('Darker Image')

plt.subplot(133)
imshow(pk_darkest)
plt.title('Darkest Image')

plt.show()

```

# Image Cropping

```python
pk_crop1 = peacock[50:(peacock.shape[0]-30), 50:(peacock.shape[1]-40)]
pk_crop2 = peacock[100:(peacock.shape[0]-40), 70:(peacock.shape[1]-30)]

#displaying image
plt.figure(1, figsize = (10,15))

plt.subplot(131)
imshow(peacock)
plt.title("Original")

plt.subplot(132)
imshow(pk_crop1)
plt.title('Cropped Image')

plt.subplot(133)
imshow(pk_crop2)
plt.title('More cropping Image')

plt.show()
```

# Edge Extraction using Sobel Filter

```python
from skimage.filters import sobel_h, sobel_v
pk_gray_sobelh= sobel_h(peacock_gray)
pk_gray_sobelv = sobel_v(peacock_gray)

#displaying image
plt.figure(1, figsize = (15,25))

plt.subplot(131)
imshow(peacock_gray)
plt.title("Original gray")

plt.subplot(132)
imshow(pk_gray_sobelh)
plt.title('Sobel_h')

plt.subplot(133)
imshow(pk_gray_sobelv)
plt.title('Sobel_v')

plt.show()
```

# Edge Extraction using Prewitt Filter

```python
from skimage.filters import prewitt_h, prewitt_v

pk_gray_prewitth= sobel_h(peacock_gray)
pk_gray_prewittv = sobel_v(peacock_gray)

#displaying image
plt.figure(1, figsize = (15,25))

plt.subplot(131)
imshow(peacock_gray)
plt.title("Original gray")

plt.subplot(132)
imshow(pk_gray_prewitth)
plt.title('Prewitt_h')

plt.subplot(133)
imshow(pk_gray_prewittv)
plt.title('Prewitt_v')

plt.show()
```
